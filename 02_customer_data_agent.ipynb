Selecting Database

```python
spark.sql("USE fraud")
```

```python
transactions = spark.sql("""
SELECT *
FROM fraud.transactions
""")

transactions.display()
```

```python
suspicious_transactions = spark.sql("""
SELECT *
FROM fraud.transactions
WHERE amount > 3000
OR risk_level = 'HIGH'
""")

suspicious_transactions.display()
```

```python
transactions_json = suspicious_transactions.toPandas().to_json(orient="records")

print(transactions_json[:1000])
```

```python
suspicious_transactions.write.format("delta").mode("overwrite").saveAsTable("fraud.suspicious_transactions")
```

```python
spark.sql("""
SELECT *
FROM fraud.suspicious_transactions
LIMIT 20
""").display()
```
